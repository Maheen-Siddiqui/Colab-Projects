```python
import numpy as np
fig, axs = plt.subplots(1, 2, figsize=(10, 6))
#creates a figure with two subplots arranged side by side, and it assigns the figure object to fig and the array of axes objects to axs.

# Subplot for Male
male_data = df[df['gender'] == 'M']
male_age_counts = male_data['Age'].value_counts().sort_index()
male_exercise_counts = male_data['exercise'].value_counts().sort_index()
axs[0].bar(np.arange(len(male_age_counts)), male_age_counts, color='skyblue', label='Age')
axs[0].bar(np.arange(len(male_exercise_counts)), male_exercise_counts, color='salmon', label='Exercise')
axs[0].set_xticks(np.arange(len(male_age_counts)))
axs[0].set_xticklabels(male_age_counts.index)
axs[0].set_xlabel('Age')
axs[0].set_ylabel('Count')
axs[0].set_title('Male: Age vs Exercise')
axs[0].legend()

# subplot for female
female_data = df[df['gender'] == 'F']
female_age_counts = female_data['Age'].value_counts().sort_index()
female_exercise_counts = female_data['exercise'].value_counts().sort_index()
axs[1].bar(np.arange(len(female_age_counts)), female_age_counts, color='skyblue', label='Age')
axs[1].bar(np.arange(len(female_exercise_counts)), female_exercise_counts, color='salmon', label='Exercise')
axs[1].set_xticks(np.arange(len(female_age_counts)))
axs[1].set_xticklabels(female_age_counts.index)
axs[1].set_xlabel('Age')
axs[1].set_ylabel('Count')
axs[1].set_title('Female: Age vs Exercise')
axs[1].legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le_smoker = LabelEncoder()
le_gender = LabelEncoder()
le_exercise = LabelEncoder()
```

```python
df['exercise_n'] = le_exercise.fit_transform(df['exercise'])
df['gender_n'] = le_gender.fit_transform(df['gender'])
df['smoker_n'] = le_smoker.fit_transform(df['smoker'])

df_n = df.drop(['exercise','gender','smoker'],axis='columns')
print(df_n.head())
```

```python
inputs = df_n.drop(['smoker_n'],axis='columns')
target = df_n.smoker_n
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(inputs,target,test_size =0.2)
print("X_train size",len(X_train))
print("X_test size",len(X_test))
```

```python
print(y_train)
```

```python
from sklearn.linear_model import LinearRegression

linear_model = LinearRegression()
linear_model.fit(X_train,y_train)
ylinear_predict = linear_model.predict(X_test)
linear_model.score(X_test,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()
logistic_model.fit(X_train,y_train)
ylogistic_predict = logistic_model.predict(X_test)
logistic_model.score(X_test,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier

d_model = DecisionTreeClassifier()
d_model.fit(X_train,y_train)
d_y_predict = d_model.predict(X_test)
d_model.score(X_test,y_test)
```

```python
from sklearn.svm import SVC

svm_model = SVC(C=10,gamma =10,kernel='poly')
svm_model.fit(X_train,y_train)
svm_y_predict = svm_model.predict(X_test)
svm_model.score(X_test,y_test)
```
